```python
#Filtragem de Audio analógica
# 2 pi rad/s = 1 hz

fs = 44100;

freq_pass = 70;


b,a  = signal.butter(8,freq_pass,btype='high', analog=True);

w, h = signal.freqs(b, a)
plt.plot(w, 20 * numpy.log10(abs(h)))
plt.xscale('log')
plt.title('Butterworth filter frequency response')
plt.xlabel('Frequency [radians / second]')
plt.ylabel('Amplitude [dB]')
plt.margins(0, 0.1)
plt.grid(which='both', axis='both')
plt.axvline(70, color='green') # cutoff frequency
plt.show()
```

```python
#Filtragem de Audio digital
# 2 pi rad/s = 1 hz

fs = sampleRate;

freq_pass = 70;

nyq = 0.5 * fs;

lowcut = 70;

highcut = 20000.0;

low = lowcut / nyq;

high = highcut / nyq;



#Ordem 8 instavel
#Ordem 7 aplicado
b,a  = signal.butter(7,low,btype='high');

w, h = signal.freqz(b, a);

freq_z = plt.figure()
plt.plot((fs * 0.5 ) * w, abs(h))
#plt.axis([0,10000])

plt.grid(which='both', axis='both')
plt.axvline(lowcut, color='green') # cutoff frequency
plt.xscale('log')
plt.show()

```

```python

h = .02  # step size in the mesh

a = numpy.logspace(-5, 3, 5)
names = []

alphas = numpy.logspace(-5, 3, 5)

for i in alphas:
    names.append('alpha ' + str(i))

classifiers = []

for i in alphas:
    classifiers.append(MLPClassifier(alpha=i, random_state=1))







list_train = numpy.asarray(range(0,10));

contadorsentences = 0;
contadorfiles = 0;
contadordire = 0;


i = 1;
k = 1;
j = 0;
m = 0;
n = 1;
# iterate over datasets
for X, y in datasets:
    figure = plt.figure(figsize = (30,225) );
    if m == 42:
        k = 2;
        n = 1;
    if contadorsentences==14:
        contadorsentences= 0;
        n+=1;
    i =0;
    for j in range(0,45):
            # preprocess dataset, split into training and test part
            X = StandardScaler().fit_transform(X)
            X_test = X[list_perm[j]];
            y_test = y[list_perm[j]];
            X_train = numpy.delete(X,list_perm[j],0);
            y_train = numpy.delete(y,list_perm[j],0);


            #X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3)


            x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
            y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
            xx, yy = numpy.meshgrid(numpy.arange(x_min, x_max, h),
                                 numpy.arange(y_min, y_max, h))

            # just plot the dataset first
            cm = plt.cm.RdBu
            cm_bright = ListedColormap(['#FF0000', '#0000FF'])
            ax = plt.subplot(45, len(classifiers) + 1, i+1)
            # Plot the training points
            ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright)
            # and testing points
            ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright, alpha=0.6)
            ax.set_xlim(xx.min(), xx.max())
            ax.set_ylim(yy.min(), yy.max())
            ax.set_xticks(())
            ax.set_yticks(())
            i += 1

            # iterate over classifiers
            for name, clf in zip(names, classifiers):
                ax = plt.subplot(45, len(classifiers) + 1, i+1)
                clf.fit(X_train, y_train)
                score = clf.score(X_test, y_test)

                # Plot the decision boundary. For that, we will assign a color to each
                # point in the mesh [x_min, x_max]x[y_min, y_max].
                if hasattr(clf, "decision_function"):
                    Z = clf.decision_function(numpy.c_[xx.ravel(), yy.ravel()])
                else:
                    Z = clf.predict_proba(numpy.c_[xx.ravel(), yy.ravel()])[:, 1]

                # Put the result into a color plot
                Z = Z.reshape(xx.shape)
                ax.contourf(xx, yy, Z, cmap=cm, alpha=.8)

                # Plot also the training points
                #ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright)
                # and testing points
                ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright,
                           alpha=0.6)

                ax.set_xlim(xx.min(), xx.max())
                ax.set_ylim(yy.min(), yy.max())
                ax.set_xticks(())
                ax.set_yticks(())
                ax.set_title(name)
                ax.text(xx.max() - .3, yy.min() + .3, ('%.2f' % score).lstrip('0'),
                        size=15, horizontalalignment='right')
                i += 1;
                

    m += 1;
    figure.subplots_adjust(left=.02, right=.98)
    if(k==1):
        plt.savefig("../Figuras/Resultados/A"+str(n)+"/sem_norma/Classificador_frase"+str(contadorsentences)+".jpg")
    else:
        plt.savefig("../Figuras/Resultados/A"+str(n)+"/com_norma/Classificador_frase"+str(contadorsentences)+".jpg")
    contadorsentences+=1;
    plt.close();
```
